```python

import matplotlib.pyplot as plt
from matplotlib.patheffects import withStroke

# Predicted results from LSTM model
predicted_democrats = 101404
predicted_republicans = 197238

# Data for the pie chart
sizes = [predicted_democrats, predicted_republicans]
colors = ['blue', 'red']  # Blue for Democrats, red for Republicans
text_colors = ['lightblue', 'lightcoral']  # Light blue for Democrats, light red for Republicans

# Create the pie chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    sizes,
    colors=colors,
    autopct='%1.2f%%',
    startangle=90,
    textprops={'fontsize': 20, 'color': 'white'},
    wedgeprops={'edgecolor': 'black'}  # Black edges around wedges
)

# Customize the percentage labels with light colors and a black border effect
for autotext, color in zip(autotexts, text_colors):
    autotext.set_color(color)  # Fill text with light color
    autotext.set_path_effects([withStroke(linewidth=3, foreground='black')])  # Add black outline to text

# Title and display
plt.title('Predicted Election Results (LSTM Model)', fontsize=16)
plt.show()
```

```python
# Data
total_voters = 101404 + 197237
registered_voters = 433332
turnout_percentage = (total_voters / registered_voters) * 100
not_voted_percentage = 100 - turnout_percentage

# Labels and data
labels = [f'Voted: {turnout_percentage:.2f}%', f'Not Voted: {not_voted_percentage:.2f}%']
sizes = [turnout_percentage, not_voted_percentage]
colors = ['green', 'grey']

# Create the bar graph
fig, ax = plt.subplots(figsize=(10, 2))
ax.barh([0], sizes[0], color=colors[0], edgecolor='black', label=labels[0], height=0.5)
ax.barh([0], sizes[1], left=sizes[0], color=colors[1], edgecolor='black', label=labels[1], height=0.5)

# Formatting
ax.set_xlim(0, 100)
ax.set_xticks([])
ax.set_yticks([])
ax.legend(loc='center', fontsize=12, frameon=False)
ax.set_title('Turnout', fontsize=14, fontweight='bold', loc='left')
plt.text(50, 0.3, "LSTM", fontsize=14, fontweight='bold', ha='right', va='center')
plt.box(False)

# Show the plot
plt.show()
```
