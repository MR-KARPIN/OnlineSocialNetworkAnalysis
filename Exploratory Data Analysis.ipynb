```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file for Florida's district data
file_path = '/content/Florida_District_all.csv'
df = pd.read_csv(file_path)

# Define the relevant district estimate columns
district_columns = [f"District {str(i).zfill(2)} Estimate" for i in range(1, 29)]

# Extract total population data from row 19 and ensure values are converted to floats
population_data = df.loc[19, district_columns].apply(lambda x: float(x.replace(",", "")))

# Sort the population data in descending order for the Pareto chart
population_data_sorted = population_data.sort_values(ascending=False)

# Calculate the cumulative percentage (if needed for other analysis)
cumulative_percentage = population_data_sorted.cumsum() / population_data_sorted.sum() * 100

# Plot the Pareto chart
fig, ax1 = plt.subplots(figsize=(12, 8))

# Bar plot for the population of each district (horizontal) with the 27th district highlighted
colors = ["orange" if idx == "District 27 Estimate" else "skyblue" for idx in population_data_sorted.index]
ax1.barh(population_data_sorted.index, population_data_sorted, color=colors)
ax1.set_ylabel('Districts')
ax1.set_xlabel('Total Population')
ax1.set_title('Pareto Chart of Population Across Florida Congressional Districts')

plt.show()
fig.savefig('total_pop.svg', format='svg', dpi=1200)

# Assuming the previously loaded and cleaned population data (population_data) is in use
# Drop the "District 28 Estimate" as it did not exist in the 2020 election
population_data_sorted = population_data.sort_values(ascending=False).drop("District 28 Estimate")

# Define the winning party for each district in 2020 (Democrat = Blue, Republican = Red)
party_wins = {
    "District 01 Estimate": "red", "District 02 Estimate": "red", "District 03 Estimate": "red",
    "District 04 Estimate": "red", "District 05 Estimate": "blue", "District 06 Estimate": "red",
    "District 07 Estimate": "blue", "District 08 Estimate": "red", "District 09 Estimate": "blue",
    "District 10 Estimate": "blue", "District 11 Estimate": "red", "District 12 Estimate": "red",
    "District 13 Estimate": "blue", "District 14 Estimate": "blue", "District 15 Estimate": "red",
    "District 16 Estimate": "red", "District 17 Estimate": "red", "District 18 Estimate": "red",
    "District 19 Estimate": "red", "District 20 Estimate": "blue", "District 21 Estimate": "blue",
    "District 22 Estimate": "blue", "District 23 Estimate": "blue", "District 24 Estimate": "blue",
    "District 25 Estimate": "red", "District 26 Estimate": "red", "District 27 Estimate": "blue"
}

# Use the color coding from the party_wins dictionary
colors = [party_wins[district] for district in population_data_sorted.index]

# Create the Pareto chart with color coding based on the election results
fig, ax1 = plt.subplots(figsize=(12, 8))

# Horizontal bar plot with party color coding
ax1.barh(population_data_sorted.index, population_data_sorted, color=colors)
ax1.set_ylabel('Districts')
ax1.set_xlabel('Total Population')
ax1.set_title('Horizontal Pareto Chart of Population Across Florida Congressional Districts\n(Colored by 2020 Presidential Election Results)')

plt.show()
fig.savefig('total_pop_colored.svg', format='svg', dpi=1200)
```

```python
# Load the dataset
file_path = '/content/Florida_District_all.csv'
df = pd.read_csv(file_path)

# Define rows of interest for exploratory data analysis
rows_of_interest = [0, 29, 200, 240, 241]
variables = ["Total population", "Hispanic %", "Mean household income", "High school +", "Bachelor degree +"]
district_columns = [f"District {str(i).zfill(2)} Estimate" for i in range(1, 28)]  # 27 districts

# Extract data for each variable and convert to appropriate numeric format
eda_data = df.loc[rows_of_interest, district_columns].applymap(lambda x: float(str(x).replace(",", "")) if isinstance(x, str) else x)
eda_data.index = variables

# Transpose dataframe for EDA, with district names and labels (presidential results)
party_wins = {
    "District 01 Estimate": "Republican", "District 02 Estimate": "Republican", "District 03 Estimate": "Republican",
    "District 04 Estimate": "Republican", "District 05 Estimate": "Democrat", "District 06 Estimate": "Republican",
    "District 07 Estimate": "Democrat", "District 08 Estimate": "Republican", "District 09 Estimate": "Democrat",
    "District 10 Estimate": "Democrat", "District 11 Estimate": "Republican", "District 12 Estimate": "Republican",
    "District 13 Estimate": "Democrat", "District 14 Estimate": "Democrat", "District 15 Estimate": "Republican",
    "District 16 Estimate": "Republican", "District 17 Estimate": "Republican", "District 18 Estimate": "Republican",
    "District 19 Estimate": "Republican", "District 20 Estimate": "Democrat", "District 21 Estimate": "Democrat",
    "District 22 Estimate": "Democrat", "District 23 Estimate": "Democrat", "District 24 Estimate": "Democrat",
    "District 25 Estimate": "Republican", "District 26 Estimate": "Republican", "District 27 Estimate": "Democrat"
}

# Convert to a transposed dataframe for EDA
eda_data_transposed = eda_data.T
eda_data_transposed['Presidential Result'] = [party_wins[district] for district in eda_data_transposed.index]

# Separate features and labels for visualization
features = eda_data_transposed[variables]
labels = eda_data_transposed['Presidential Result']

# Visualize distributions by presidential result
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Florida Congressional Districts Boxplot by Presidential Result')

# Total Population Distribution
sns.boxplot(x=labels, y=features["Total population"], ax=axes[0, 0], palette={"Republican": "red", "Democrat": "blue"})
axes[0, 0].set_title("Total Population")

# Hispanic % Distribution
sns.boxplot(x=labels, y=features["Hispanic %"], ax=axes[0, 1], palette={"Republican": "red", "Democrat": "blue"})
axes[0, 1].set_title("Hispanic %")

# Mean Household Income Distribution
sns.boxplot(x=labels, y=features["Mean household income"], ax=axes[1, 0], palette={"Republican": "red", "Democrat": "blue"})
axes[1, 0].set_title("Mean Household Income")

# High School+ Education Distribution
sns.boxplot(x=labels, y=features["High school +"], ax=axes[1, 1], palette={"Republican": "red", "Democrat": "blue"})
axes[1, 1].set_title("High School+ Education")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
fig.savefig('boxplot1.svg', format='svg', dpi=1200)

```

```python
# Load the CSV file
file_path = '/content/countypres_2000-2020.csv'  # Replace with your actual file path
election_data = pd.read_csv(file_path)

# Filter data for Miami-Dade County for the US Presidential elections
miami_dade_data = election_data[(election_data['county_name'] == 'MIAMI-DADE') & (election_data['office'] == 'US PRESIDENT')]

# Define custom colors for each party
party_colors = {
    'LIBERTARIAN': 'yellow',
    'REPUBLICAN': 'red',
    'DEMOCRAT': 'blue',
    'OTHER': 'grey',
    'GREEN': 'green'
}

# --- First Plot: Absolute Votes ---

# Group the Miami-Dade data by year and party to calculate total votes per party per year
miami_dade_results = miami_dade_data.groupby(['year', 'party'])['candidatevotes'].sum().unstack()

# Plotting the absolute votes
plt.figure(figsize=(12, 6))
miami_dade_results.plot(kind='line', marker='o', color=[party_colors.get(col, 'black') for col in miami_dade_results.columns], figsize=(12, 6))
plt.title('US Presidential Election Results in Miami-Dade County (2000-2020) - Absolute Votes')
plt.xlabel('Year')
plt.ylabel('Votes')
plt.legend(title='Party')
plt.xticks([2000,2004,2008,2012,2016,2020])
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Second Plot: Percentage of Votes ---

# Calculate the percentage of votes per party by dividing candidatevotes by totalvotes
miami_dade_data['percentage_votes'] = (miami_dade_data['candidatevotes'] / miami_dade_data['totalvotes']) * 100

# Group by year and party to get the average percentage for each party per year
miami_dade_percentage = miami_dade_data.groupby(['year', 'party'])['percentage_votes'].sum().unstack()

# Plotting the percentage results
plt.figure(figsize=(12, 6))
miami_dade_percentage.plot(kind='line', marker='o', color=[party_colors.get(col, 'black') for col in miami_dade_percentage.columns], figsize=(12, 6))
plt.title('US Presidential Election Results in Miami-Dade County (2000-2020) - Percentage of Total Votes')
plt.xlabel('Year')
plt.ylabel('Percentage of Votes (%)')
plt.legend(title='Party')
plt.xticks([2000,2004,2008,2012,2016,2020])
plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
# Load the dataset
file_path = '/content/1976-2022-house.csv'
df = pd.read_csv(file_path)

# Filter data for Florida's 27th district from 2012 to 2022
florida_district_27 = df[(df['year'] >= 2012) & (df['year'] <= 2022) & (df['district'] == 27) & (df['state'] == 'FLORIDA')]

# Remove data for the year 2014
florida_district_27 = florida_district_27[florida_district_27['year'] != 2014]

# Separate Republican and Democrat votes
republican_votes = florida_district_27[florida_district_27['party'] == 'REPUBLICAN']
democrat_votes = florida_district_27[florida_district_27['party'] == 'DEMOCRAT']

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(republican_votes['year'], republican_votes['candidatevotes'], marker='o', color='red', label='Republican')
plt.plot(democrat_votes['year'], democrat_votes['candidatevotes'], marker='o', color='blue', label='Democrat')

plt.xlabel('Year')
plt.ylabel('Candidate Votes')
plt.title('House of Representatives Votes in Florida\'s 27th District (2012-2022, excluding 2014)')
plt.legend()
plt.grid(True)
plt.show()


# Calculate vote percentages
florida_district_27['vote_percentage'] = florida_district_27['candidatevotes'] / florida_district_27['totalvotes']

# Separate Republican and Democrat votes with percentages
republican_votes_pct = florida_district_27[(florida_district_27['party'] == 'REPUBLICAN')]
democrat_votes_pct = florida_district_27[(florida_district_27['party'] == 'DEMOCRAT')]

# Create the plot with percentages
plt.figure(figsize=(10, 6))
plt.plot(republican_votes_pct['year'], republican_votes_pct['vote_percentage'], marker='o', color='red', label='Republican')
plt.plot(democrat_votes_pct['year'], democrat_votes_pct['vote_percentage'], marker='o', color='blue', label='Democrat')

plt.xlabel('Year')
plt.ylabel('Vote Percentage')
plt.title('House of Representatives Vote Percentage in Florida\'s 27th District (2012-2022, excluding 2014)')
plt.legend()
plt.grid(True)
plt.show()
```
