```python
# Convert 'Year' column in turnoutdf to int64
turnoutdf['Year'] = turnoutdf['Year'].astype(int)  # Convert to int64

# Merge turnoutdf and censusdf based on 'Year'
mergedTurnout_df = pd.merge(turnoutdf, df_other_years, on='Year', how='inner')
```

```python
# Split the 'Miami-Dade Turnout' feature
miami_dade_turnout = mergedTurnout_df['Miami-Dade Turnout']
mergedTurnout_df.drop(columns=['Miami-Dade Turnout'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Fit and transform the 'ElectionType' column
mergedTurnout_df['ElectionTypeEncoded'] = label_encoder.fit_transform(mergedTurnout_df['ElectionType'])
```

```python
# Drop the original columns that were preprocessed (if they exist)
if 'ElectionType' in mergedTurnout_df.columns:
    mergedTurnout_df = mergedTurnout_df.drop(columns=['ElectionType'])


print(mergedTurnout_df.head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score


# Prepare the features (X) and target variable (y)
mergedTurnout_df.fillna(0, inplace=True)
X = mergedTurnout_df
y = miami_dade_turnout

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Define the models
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(random_state=42),  # Added random_state for reproducibility
    "SVM": SVR(),
    "KNN": KNeighborsRegressor()
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = {"MSE": mse, "R-squared": r2}

# Print the results
for name, metrics in results.items():
    print(f"Model: {name}")
    print(f"  Mean Squared Error: {metrics['MSE']}")
    print(f"  R-squared: {metrics['R-squared']}")
```

```python
pre_data = pre_data.drop(columns=['Juris1num'], errors='ignore')
pre_data_summed = pre_data.groupby(['Year', 'PartyCode'])['CanVotes'].sum().reset_index()
pre_data_summed
```

```python
# Create a dictionary to store the yearly sums
yearly_sums = {}

# Iterate through the rows of the DataFrame
for index, row in pre_data_summed.iterrows():
    year = row['Year']
    party = row['PartyCode']
    votes = row['CanVotes']

    # If the year is not in the dictionary, create a new entryDICTION

    if year not in yearly_sums:
        yearly_sums[year] = {'REP': 0, 'DEM': 0, 'OTHER': 0}

    # Update the vote counts based on the party
    if party == 'REP':
        yearly_sums[year]['REP'] += votes
    elif party == 'DEM':
        yearly_sums[year]['DEM'] += votes
    else:
        yearly_sums[year]['OTHER'] += votes

# Create a new DataFrame from the yearly sums
new_df = pd.DataFrame.from_dict(yearly_sums, orient='index')

# Reset the index and rename columns
new_df = new_df.reset_index().rename(columns={'index': 'Year'})

new_df
```

```python
import matplotlib.pyplot as plt

# Assuming 'new_df' is your DataFrame with 'Year', 'REP', and 'DEM' columns

plt.figure(figsize=(10, 6))  # Adjust figure size as needed

plt.plot(new_df['Year'], new_df['REP'], label='Republican Votes', marker='o')
plt.plot(new_df['Year'], new_df['DEM'], label='Democrat Votes', marker='x')

plt.xlabel('Year')
plt.ylabel('Number of Votes')
plt.title('Tendency of REP and DEM Votes Over the Years')
plt.legend()
plt.grid(True)  # Add a grid for better readability
plt.show()
```
